# GridTraveller Unmemoized

```python
def grid_traveller(n: int, m: int) -> int:
    if n == 0 or m == 0:
        return 0
    if n == 1 and m == 1:
        return 1
    return grid_traveller(n - 1, m) + grid_traveller(n, m - 1)
```

```python
%%time
grid_traveller(20, 10)
```

# GridTraveller Memoized

```python
def grid_traveller_memo(n: int, m: int, memo={}) -> int:
    if n == 0 or m == 0:
        return 0
    if n == 1 and m == 1:
        return 1
    if f"{n},{m}" in memo: return memo[f"{n},{m}"]
    memo[f"{n},{m}"] = grid_traveller_memo(n - 1, m, memo) + grid_traveller_memo(n, m - 1, memo)
    return memo[f"{n},{m}"]
```

```python
%%time
grid_traveller_memo(18, 18)
```
