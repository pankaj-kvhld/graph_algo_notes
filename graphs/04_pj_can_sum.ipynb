# CanSum Unmemoized

```python
def can_sum(target: int, nums: list) -> bool:
    if target == 0:
        return True
    if target < 0:
        return False

    for num in nums:
        if can_sum(target - num, nums):
            return True
        
    return False
```

```python
can_sum(7, [5, 3, 4, 7])
```

```python
can_sum(7, [2, 4, 8])
```

```python
%%time
can_sum(251, [7, 14])
```

## Complexity

- Time complexity `O(n^m)` where `n` is the target and `m` is the length of `nums`.
- Space complexity is `O(n)`

# CanSum Memoized

```python
def can_sum(target: int, nums: list, memo={}) -> bool:
    if target == 0:
        return True
    if target < 0:
        return False
    if target in memo:
        return memo[target]

    for num in nums:
        if can_sum(target - num, nums, memo):
            memo[target - num] = True
            return True

    memo[target] = False
    return False
```

## Complexity

- Time complexity `O(n*m)` where `n` is the target and `m` is the length of `nums`.
- Space complexity is `O(n)`

```python
can_sum(7, [5, 3, 4, 7])
```

```python
can_sum(7, [2, 4, 8])
```

```python
%%time
can_sum(300, [7, 14])
```
